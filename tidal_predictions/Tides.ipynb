Compare tidal predictions calculated with sythesis flag and without. Synthesis flag removes tidal constituents with low signal to noise ratio.

```python
#synthesis 2
tfile='Point Atkinson_t_tide_compare8_31-Dec-2013_02-Dec-2015.csv'
ttide1,msl= stormtools.load_tidal_predictions(tfile)
diff1=ttide1.pred_all-ttide1.pred_8

#no synthesis
tfile='/data/nsoontie/MEOPAR/analysis/Nancy/tides/Point Atkinson_t_tide_compare8_31-Dec-2013_02-Dec-2015.csv'
ttide2,msl= stormtools.load_tidal_predictions(tfile)
diff2=ttide2.pred_all-ttide2.pred_8
```

```python
%matplotlib inline
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,6))

ax.plot(ttide2.time,ttide2.pred_all,'o-',label='no synthesis')
ax.plot(ttide1.time,ttide1.pred_all,'o-',label='synthesis 2')
ax.plot(ttide1.time,ttide1.pred_8,'o',label='pred 8')
ax.set_xlim([datetime.datetime(2014,12,20),datetime.datetime(2014,12,31)])
ax.grid()
ax.legend(loc=0)
```

No synthesis (blue) is slightly lower, but only at certain times of the tidal cycle (peak). This suggests to me that I need to be more careful with my tidal precitions and the corrections.
